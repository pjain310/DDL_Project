```python
image = cv2.cvtColor(cv2.imread("/Users/prerna/Downloads/607216-gettyimages-1181927942.jpg"), cv2.COLOR_RGB2GRAY)
#image = cv2.cvtColor(cv2.imread("data_subset/test/0110417.jpg"), cv2.COLOR_RGB2GRAY)
sobelx = cv2.Sobel(image,cv2.CV_64F, 1, 0, 5)
sobely = cv2.Sobel(image,cv2.CV_64F, 0, 1, 5)
scharrx = cv2.Scharr(image,cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(image,cv2.CV_64F, 0, 1)
laplacian = cv2.Laplacian(image,cv2.CV_64F)
equ = cv2.equalizeHist(image)
```

```python
plt.figure(figsize=(22,12))
plt.subplot(332), plt.imshow(image,cmap='gray'),plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(334), plt.imshow(sobelx,cmap='gray'),plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(335), plt.imshow(sobely,cmap='gray'),plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(337), plt.imshow(scharrx,cmap='gray'),plt.title('Scharr X'), plt.xticks([]), plt.yticks([])
plt.subplot(338), plt.imshow(scharry,cmap='gray'),plt.title('Scharr Y'), plt.xticks([]), plt.yticks([])
plt.subplot(336), plt.imshow(laplacian,cmap='gray'),plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(339), plt.imshow(equ,cmap='gray'),plt.title('Equalized Histogram'), plt.xticks([]), plt.yticks([])
plt.show()
```
